```python
from bs4 import BeautifulSoup
import requests
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize 
from nltk.stem import PorterStemmer
```

Q1 fetch the text content


```python
URL = "https://webscraper.io/test-sites/e-commerce/allinone"
TESLA = "https://en.wikipedia.org/wiki/Tesla,_Inc."
MEDIUM = "https://uxdesign.cc/key-takeaways-from-airbnbs-winter-redesign-c1de0efa7818"
page = requests.get(MEDIUM).content
soup = BeautifulSoup(page, 'html.parser')
```

# GETTING PARA FROM BODY

```python
text_corpus = []
for tags in soup.stripped_strings:
    test = repr(tags)
    print(test)
    text_corpus.append(test)
```

# CLEANING STARTS


```python
file = open("clean-text.txt","w")
additional_chars = ['[', ']', '\\n', '|', '\\t', "'", "' '","''", ',', '(',')','``',':','-','--','"','_','—',';']
for idx in range(len(text_corpus)):
    text = text_corpus[idx]
    for removal_chars in additional_chars:
        text = text.replace(removal_chars, " ")
        text = text.strip()
        text = " ".join(text.split())
        text_corpus[idx] = text
        
    file.write(text+"\n")
print(text_corpus)
file.close()
```

# GET ALL LINKS

```python
LINKS= ["src","href","url"]
all_links = []    
for attr in LINKS:
    tags = soup.select(f"[{attr}]")
    all_links += [tag.attrs[attr] for tag in tags]

for i in range(len(all_links)):
    print(i+1, all_links[i])
```

# TOKENIZATION

```python
tokens = []
for line in text_corpus:
    tokens+=word_tokenize(line)
for token in tokens:
    print(token)
```

# UNIQUE WORDS

```python
unique_words = sorted(set(tokens))
for uw in unique_words:
    print(uw)
```

# STEMMED WORDS

```python
stemmed_words = []
ps = PorterStemmer()
for word in unique_words:
   stemmed_words.append(ps.stem(word))
for idx in range(len(stemmed_words)):
    print(idx+1, stemmed_words[idx])
```

# SENTENCES

```python
sentences = []
for line in text_corpus:
    sentence = sent_tokenize(line)
    sentences+=sentence
for sentence in sentences:
    print(sentence)
```
